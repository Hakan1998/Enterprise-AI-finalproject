```python
from zenml.client import Client

# Initialize ZenML client
client = Client()

# Get all artifacts
artifacts = client.list_artifacts()

# Print all artifacts
for artifact in artifacts:
    print(f"Artifact ID: {artifact.id}")
    print(f"Artifact Name: {artifact.name}")
```
